```python
taxonomy_table = pd.read_csv(
    "/home/jose.gomez/Projects/16S_completo/test/orina/RESDIR/taxonomy-table.tsv",
    sep="\t",
    header=0,
    index_col=0
    )
taxonomy_table
```

```python
def create_index_dict(df):
    max_rows = len(df)
    max_digits = len(str(max_rows))
    old_to_new_index_map = {}
    for i, old_index in enumerate(df.index):
        new_index = f"seq_{str(i+1).zfill(max_digits)}"
        old_to_new_index_map[old_index] = new_index
    return old_to_new_index_map
```

```python
id_map = create_index_dict(count_table)
```

```python
count_table = count_table.rename(index=id_map)
count_table
```

```python
taxonomy_table = taxonomy_table.rename(index=id_map)
taxonomy_table
```

```python
taxonomy_table["lineage"] = taxonomy_table["lineage"].str.replace("\\w__", "", regex=True)
taxonomy_table = taxonomy_table["lineage"].str.split(";", expand=True)
taxonomy_table.columns = ["Root", "Kingdom", "Phylum", "Class", "Order", "Family", "Genus", "Species"]
taxonomy_table
```

```python
taxonomy_table.to_csv(
    "/home/jose.gomez/Projects/16S_completo/test/orina/RESDIR/taxonomy_table_parsed.tsv",
    sep="\t",
    header=True,
    index=True,
)
```
